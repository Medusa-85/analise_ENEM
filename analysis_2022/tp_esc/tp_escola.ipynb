## Gr√°ficos

```python
import matplotlib.pyplot as plt
plt.style.use('_mpl-gallery')
```

### **GR√ÅFICOS - TP_ESCOLA**

```python
df[["MEDIA_GERAL", "TP_ESCOLA"]].groupby("TP_ESCOLA").describe()
```

```python
grouped_stats = df[["MEDIA_GERAL", "TP_ESCOLA"]].groupby("TP_ESCOLA").describe()

# Cria√ß√£o de um gr√°fico separado para cada medida estat√≠stica
for stat in ["mean", "min", "max", "count"]:
    fig, ax = plt.subplots(figsize=(8, 6))
    
    # Extra√ß√£o dos dados para o gr√°fico
    x_labels = grouped_stats.index
    y_values = grouped_stats[("MEDIA_GERAL", stat)]
    
    # Plotagem do gr√°fico
    ax.bar(x_labels, y_values, align='center', alpha=0.7)
    
    # Configura√ß√µes do gr√°fico
    ax.set_title(f'{stat.capitalize()} por Tipo de Escola')
    ax.set_xlabel('Tipo de Escola')
    ax.set_ylabel(f'{stat.capitalize()} da M√©dia Geral')
    ax.set_xticks(x_labels)
    ax.set_xticklabels(x_labels)
    ax.grid(True)
    
    # Mostrar o gr√°fico
    plt.tight_layout()
    plt.show()
```

### **GR√ÅFICOS - TP_LOCALIZACAO_ESC**

```python
df[["MEDIA_GERAL", "TP_LOCALIZACAO_ESC"]].groupby("TP_LOCALIZACAO_ESC").describe()
```

```python
grouped_stats = df[["MEDIA_GERAL", "TP_LOCALIZACAO_ESC"]].groupby("TP_LOCALIZACAO_ESC").describe()

# Cria√ß√£o de um gr√°fico separado para cada medida estat√≠stica
for stat in ["mean", "min", "max", "count"]:
    fig, ax = plt.subplots(figsize=(8, 6))
    
    # Extra√ß√£o dos dados para o gr√°fico
    x_labels = grouped_stats.index
    y_values = grouped_stats[("MEDIA_GERAL", stat)]
    
    # Plotagem do gr√°fico
    ax.bar(x_labels, y_values, align='center', alpha=0.7)
    
    # Configura√ß√µes do gr√°fico
    ax.set_title(f'{stat.capitalize()} por Tipo de Escola')
    ax.set_xlabel('Tipo de Escola')
    ax.set_ylabel(f'{stat.capitalize()} da M√©dia Geral')
    ax.set_xticks(x_labels)
    ax.set_xticklabels(x_labels)
    ax.grid(True)
    
    # Mostrar o gr√°fico
    plt.tight_layout()
    plt.show()
```

### **GR√ÅFICOS - TP_ESCOLA e TP_LOCALIZACAO_ESC**

```python
df[["MEDIA_GERAL", "TP_ESCOLA", "TP_LOCALIZACAO_ESC"]].groupby(["TP_ESCOLA", "TP_LOCALIZACAO_ESC"]).describe()
```

```python
# Realiza√ß√£o do agrupamento e c√°lculo das estat√≠sticas
grouped = df.groupby(["TP_ESCOLA", "TP_LOCALIZACAO_ESC"])[["MEDIA_GERAL"]].agg(['mean', 'max', 'min', 'count'])

# Reinicializa√ß√£o os √≠ndices para poder acessar as colunas de agrupamento
grouped.reset_index(inplace=True)

# Cria√ß√£o de uma lista de r√≥tulos para os grupos
labels = [f"Escola {tp_escola}, Localiza√ß√£o {tp_localizacao}" for tp_escola, tp_localizacao in zip(grouped["TP_ESCOLA"], grouped["TP_LOCALIZACAO_ESC"])]

# Itera√ß√£o sobre as estat√≠sticas e gere gr√°ficos individuais
statistics = ['mean', 'max', 'min', 'count']
for stat in statistics:
    plt.figure(figsize=(10, 6))
    plt.bar(labels, grouped[("MEDIA_GERAL", stat)])
    plt.xticks(rotation=45, ha="right")
    plt.xlabel("Combina√ß√£o de Tipo de Escola e Localiza√ß√£o")
    plt.ylabel(stat.capitalize())  # Define o r√≥tulo do eixo Y com a estat√≠stica atual
    plt.title(f"{stat.capitalize()} por Combina√ß√£o de Tipo de Escola e Localiza√ß√£o")
    plt.tight_layout()

    plt.show()
```
