## **Estat√≠sticas da vari√°vel: Tipo da escola do Ens. M√©dio (TP_ESCOLA)**

*Dicion√°rio - Base de dados do governo:*

1 - N√£o respondeu

2 - P√∫blica

3 - Privada

Geral

```python
media_geral = df["TP_ESCOLA"].mean()
minimo_geral = df["TP_ESCOLA"].min()
maximo_geral = df["TP_ESCOLA"].max()
mediana_geral = df["TP_ESCOLA"].median()

print(f"AN√ÅLISE DESCRITIVA GERAL: \nM√©dia: {media_geral}\nM√≠nimo: {minimo_geral}\nM√°ximo: {maximo_geral}\nMediana: {mediana_geral}")
```

```python
df.columns
```

```python
df["MEDIA_GERAL"] = df[["NU_NOTA_CN", "NU_NOTA_CH", "NU_NOTA_LC", "NU_NOTA_MT", "NU_NOTA_REDACAO"]].mean(axis=1)
df.to_csv('enem_2022_v2.csv')

df.head()
```
