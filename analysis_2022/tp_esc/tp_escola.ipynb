## Gráficos

```python
import matplotlib.pyplot as plt
plt.style.use('_mpl-gallery')
```

### **GRÁFICOS - TP_ESCOLA**

```python
df[["MEDIA_GERAL", "TP_ESCOLA"]].groupby("TP_ESCOLA").describe()
```

```python
grouped_stats = df[["MEDIA_GERAL", "TP_ESCOLA"]].groupby("TP_ESCOLA").describe()

# Criação de um gráfico separado para cada medida estatística
for stat in ["mean", "min", "max", "count"]:
    fig, ax = plt.subplots(figsize=(8, 6))
    
    # Extração dos dados para o gráfico
    x_labels = grouped_stats.index
    y_values = grouped_stats[("MEDIA_GERAL", stat)]
    
    # Plotagem do gráfico
    ax.bar(x_labels, y_values, align='center', alpha=0.7)
    
    # Configurações do gráfico
    ax.set_title(f'{stat.capitalize()} por Tipo de Escola')
    ax.set_xlabel('Tipo de Escola')
    ax.set_ylabel(f'{stat.capitalize()} da Média Geral')
    ax.set_xticks(x_labels)
    ax.set_xticklabels(x_labels)
    ax.grid(True)
    
    # Mostrar o gráfico
    plt.tight_layout()
    plt.show()
```

### **GRÁFICOS - TP_LOCALIZACAO_ESC**

```python
df[["MEDIA_GERAL", "TP_LOCALIZACAO_ESC"]].groupby("TP_LOCALIZACAO_ESC").describe()
```

```python
grouped_stats = df[["MEDIA_GERAL", "TP_LOCALIZACAO_ESC"]].groupby("TP_LOCALIZACAO_ESC").describe()

# Criação de um gráfico separado para cada medida estatística
for stat in ["mean", "min", "max", "count"]:
    fig, ax = plt.subplots(figsize=(8, 6))
    
    # Extração dos dados para o gráfico
    x_labels = grouped_stats.index
    y_values = grouped_stats[("MEDIA_GERAL", stat)]
    
    # Plotagem do gráfico
    ax.bar(x_labels, y_values, align='center', alpha=0.7)
    
    # Configurações do gráfico
    ax.set_title(f'{stat.capitalize()} por Tipo de Escola')
    ax.set_xlabel('Tipo de Escola')
    ax.set_ylabel(f'{stat.capitalize()} da Média Geral')
    ax.set_xticks(x_labels)
    ax.set_xticklabels(x_labels)
    ax.grid(True)
    
    # Mostrar o gráfico
    plt.tight_layout()
    plt.show()
```

### **GRÁFICOS - TP_ESCOLA e TP_LOCALIZACAO_ESC**

```python
df[["MEDIA_GERAL", "TP_ESCOLA", "TP_LOCALIZACAO_ESC"]].groupby(["TP_ESCOLA", "TP_LOCALIZACAO_ESC"]).describe()
```

```python
# Realização do agrupamento e cálculo das estatísticas
grouped = df.groupby(["TP_ESCOLA", "TP_LOCALIZACAO_ESC"])[["MEDIA_GERAL"]].agg(['mean', 'max', 'min', 'count'])

# Reinicialização os índices para poder acessar as colunas de agrupamento
grouped.reset_index(inplace=True)

# Criação de uma lista de rótulos para os grupos
labels = [f"Escola {tp_escola}, Localização {tp_localizacao}" for tp_escola, tp_localizacao in zip(grouped["TP_ESCOLA"], grouped["TP_LOCALIZACAO_ESC"])]

# Iteração sobre as estatísticas e gere gráficos individuais
statistics = ['mean', 'max', 'min', 'count']
for stat in statistics:
    plt.figure(figsize=(10, 6))
    plt.bar(labels, grouped[("MEDIA_GERAL", stat)])
    plt.xticks(rotation=45, ha="right")
    plt.xlabel("Combinação de Tipo de Escola e Localização")
    plt.ylabel(stat.capitalize())  # Define o rótulo do eixo Y com a estatística atual
    plt.title(f"{stat.capitalize()} por Combinação de Tipo de Escola e Localização")
    plt.tight_layout()

    plt.show()
```
