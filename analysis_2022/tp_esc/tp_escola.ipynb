## **Estatísticas da variável: Tipo da escola do Ens. Médio (TP_ESCOLA)**

*Dicionário - Base de dados do governo:*

1 - Não respondeu

2 - Pública

3 - Privada

Geral

```python
media_geral = df["TP_ESCOLA"].mean()
minimo_geral = df["TP_ESCOLA"].min()
maximo_geral = df["TP_ESCOLA"].max()
mediana_geral = df["TP_ESCOLA"].median()

print(f"ANÁLISE DESCRITIVA GERAL: \nMédia: {media_geral}\nMínimo: {minimo_geral}\nMáximo: {maximo_geral}\nMediana: {mediana_geral}")
```

Específica

```python
df.columns
```

```python
df["MEDIA_GERAL"] = df[["NU_NOTA_CN", "NU_NOTA_CH", "NU_NOTA_LC", "NU_NOTA_MT", "NU_NOTA_REDACAO"]].mean(axis=1)
df.to_csv('enem_2022_v2.csv')

df.head()
```

```python
df[["MEDIA_GERAL", "TP_ESCOLA"]].groupby("TP_ESCOLA").describe()
```
