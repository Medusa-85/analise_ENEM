## Média das Notas por ocupação do pai

A ocupação do pai é um fator socioeconômico importante que pode estar correlacionado com a nota do aluno no ENEM. Aqui estão algumas considerações sobre a possível correlação entre a ocupação do pai e a nota no ENEM, com base nos grupos de ocupações que você forneceu:

Acesso a Recursos Educacionais e Culturais: A ocupação do pai pode influenciar o nível de renda da família e, por sua vez, o acesso a recursos educacionais e culturais. Pais em ocupações dos grupos de maior nível socioeconômico (por exemplo, Grupo 5) podem ter mais recursos para investir na educação de seus filhos, incluindo aulas particulares, materiais de estudo e experiências culturais enriquecedoras.

Motivação e Valorização da Educação: Em geral, pais com ocupações mais altas tendem a valorizar a educação e podem incentivar seus filhos a se dedicarem aos estudos. Isso pode criar um ambiente propício para o aprendizado e melhor desempenho acadêmico.

Disponibilidade de Tempo e Suporte: Pais com ocupações que demandam menos tempo de trabalho ou têm horários mais flexíveis podem ter mais tempo para oferecer suporte aos estudos de seus filhos. Isso pode incluir ajudar nas lições de casa, tirar dúvidas e fornecer apoio emocional.

Experiências Anteriores e Nível Educacional: Pais com ocupações que exigem um alto nível de educação (por exemplo, Grupo 5) podem ter uma maior compreensão dos desafios educacionais e podem estar mais bem equipados para oferecer orientação acadêmica aos filhos.

Nível de Estresse e Ambiente Familiar: O estresse associado a ocupações de alta responsabilidade ou demanda pode afetar o ambiente familiar e, consequentemente, a disposição e a capacidade do aluno de se concentrar nos estudos.

Ambiente de Aprendizado em Casa: A ocupação do pai também pode afetar o ambiente de aprendizado em casa. Pais com ocupações que envolvem habilidades técnicas (por exemplo, Grupo 3) podem estar mais propensos a ajudar com tarefas relacionadas a ciências e matemática, enquanto pais com ocupações mais intelectuais (por exemplo, Grupo 4) podem ajudar em áreas como humanidades e idiomas.

# Ocupação do pai vs média de todas as notas dos alunos:

```python
# Calcular a média das notas das colunas especificadas
df_enem_2022['media_notas'] = df_enem_2022[['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_REDACAO']].mean(axis=1)

# Mapeamento das descrições para os códigos de Q001
q003_labels = {
    'A': 'Lavrador, Boia Fria, pescador... ',
    'B': 'Diarista, Vigilante, Porteiro...',
    'C': 'Padeiro, Cozinheiro, Operador de Maquinas...',
    'D': 'Professor, Microempresario, Autonomo...',
    'E': 'Medico, engenheiro, dentista...',
    'F': 'Não sabe.'
}

# Seu código para criar o gráfico de barras
sns.set(style="whitegrid")  # Estilo do gráfico
plt.figure(figsize=(10, 6))  # Tamanho da figura

# Substitua 'df_2022' pelo seu DataFrame
ax = sns.barplot(x='Q003', y='media_notas', data=df_enem_2022)

# Personalização dos rótulos do eixo X
ax.set_xticklabels([q003_labels[code] for code in sorted(q003_labels.keys())], rotation=45, ha='right')

# Configurações adicionais
plt.xlabel('Ocupações (Q003)')
plt.ylabel('Soma das Notas')
plt.title('media_notas por ocupação do pai')
plt.tight_layout() 


# Exibir o gráfico
plt.show()
```
