```python
# Calcular a média da soma das notas das colunas especificadas
df_enem_2022['soma_notas'] = df_enem_2022[['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_REDACAO']].sum(axis=1)


media = df_enem_2022['soma_notas'].mean()
maximo = df_enem_2022['soma_notas'].max()
minimo = df_enem_2022['soma_notas'].min()
mediana = df_enem_2022['soma_notas'].median()

print(f'O máximo da soma das notas é {round(maximo, 2)}')
print(f'A média da soma das notas é {round(media, 2)}')
print(f'A mediana da soma das notas é {round(mediana, 2)}')
print(f'O mínimo da soma das notas é {round(minimo, 2)}')
```

```python
colunas_presenca = ['TP_PRESENCA_CN', 'TP_PRESENCA_CH', 'TP_PRESENCA_LC', 'TP_PRESENCA_MT']

# Filtra as linhas onde todas as colunas de presença possuem valor 0
linhas_presenca_0 = df_enem_2022[(df_enem_2022[colunas_presenca] == 0).all(axis=1)]

# Contendo contagem do que acreditamos ser dos alunos que faltaram a prova.
print(len(linhas_presenca_0))
```
