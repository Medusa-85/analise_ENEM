### **Alteração dos dados**

Alteração dos dados para redução do dicionário.

De acordo com definição de classes do IBGE:

Classe A - Acima de 20 salários mínimos - 24.240,01 ou mais 
-- corresponde a: (Q)

Classe B - De 10 a 20 salários mínimos - 12.120,01 a 22.240,00
-- corresponde a: (N, O, P)

Classe C - De 4 a 10 salários mínimos - 4.848,01 a 12.120,00 
-- corresponde a: (G, H, I, J, K, L, M)

Classe D - De 2 a 4 salários mínimos - 2.424,01 a 4.848,00 
-- corresponde a: (E, F)

Classe E - Até 2 salários mínimos - Até 2.424,00 
-- corresponde a: (A, B, C, D)

```python
df['CLASSES'] = df['Q006'].apply(lambda x: 'A' if x == 'Q' 
                                 else 'B' if x == 'N' or x == 'O' or x == 'P' 
                                 else 'C' if x == 'G' or x == 'H' or x == 'I' or x == 'J' or x == 'K' or x == 'L' or x == 'M'
                                 else 'D' if x == 'E' or x == 'F'
                                 else 'E')
```

### Estatísticas dos dados

```python
df["MEDIA_GERAL"] = df[["NU_NOTA_CN", "NU_NOTA_CH", "NU_NOTA_LC", "NU_NOTA_MT", "NU_NOTA_REDACAO"]].mean(axis=1)

df[["MEDIA_GERAL", "CLASSES"]].groupby("CLASSES").describe()
```

## Gráficos

```python
import matplotlib.pyplot as plt
plt.style.use('_mpl-gallery')
```

```python
df[["MEDIA_GERAL", "CLASSES"]].groupby("CLASSES").describe()
```

```python
grouped_stats = df[["MEDIA_GERAL", "CLASSES"]].groupby("CLASSES").describe()

# Criação de um gráfico separado para cada medida estatística
for stat in ["mean", "min", "max", "count"]:
    fig, ax = plt.subplots(figsize=(8, 6))
    
    # Extração dos dados para o gráfico
    x_labels = grouped_stats.index
    y_values = grouped_stats[("MEDIA_GERAL", stat)]
    
    # Plotagem do gráfico
    ax.bar(x_labels, y_values, align='center', alpha=0.7)
    
    # Configurações do gráfico
    ax.set_title(f'{stat.capitalize()} por Classe')
    ax.set_xlabel('Classes')
    ax.set_ylabel(f'{stat.capitalize()} da Média Geral')
    ax.set_xticks(x_labels)
    ax.set_xticklabels(x_labels)
    ax.grid(True)
    
    # Mostrar o gráfico
    plt.tight_layout()
    plt.show()
```
